```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("C:\Users\Akul Chhillar\Downloads\House Price Data.xlsx")
```

```python
df.drop(["id","date"],axis=1,inplace=True)
```

```python
range = df.price.max()-df.price.min()
```

```python
sns.countplot(df.bedrooms)
```

We have maximum houses having 3 bedrooms

We can see that we have a strong correlation between Square Foot and the price of house

```python
plt.scatter(x=df.bedrooms,y=df.price)
print df[["bedrooms","price"]].corr().iloc[:,1][0]
```

We have a correlation of 0.38 between No. of bedrooms and prices of houses . The correlation is not strong even as depicted by the graph.

```python
plt.scatter(df.price,df.sqft_living)
print df[["sqft_living","price"]].corr().iloc[:,1][0]
```

We have a correlation of 0.7 between square foot and prices of houses . 

```python
plt.scatter((df['sqft_living']+df['sqft_basement']),df['price'])
```

Over here we see a strong correlation between living space and price of a house. this is quite intuitive.

```python
sns.countplot(x=df.grade,data=df)
```

We have maximum of grade 7 houses.

```python
sns.countplot(x=df.condition,data=df)
```

Condition 3 houses are maximum in dataset.

```python
corr = df[["price","bedrooms","sqft_living","floors","grade","condition"]].corr()
sns.heatmap(data=corr,annot=True,cbar=True)
```

It is a better way to look at the data. If we look at the left most column we will find the correlation of various elements with prices.
